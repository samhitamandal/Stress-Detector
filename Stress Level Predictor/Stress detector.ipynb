# **Data Preprocessing**

```python
df["Stress Level"].unique()
```

```python
str = df.loc[:, 'Stress Level']

str[ str<4 ] = 0
str[ (3<str) & (str<=6) ] = 1
str[ (6<str) & (str<=10) ] = 2
```

```python
df
```

```python
df=pd.get_dummies(df, columns=["Occupation"],dtype=int)
dict={"Male":1,
         "Female":0
       }
df['Gender']=df['Gender'].map(dict)
df.head()
```

```python
df[['Systolic Pressure', 'Diastolic Pressure']] = df['Blood Pressure'].str.split('/', expand = True)
```

```python
df
```

```python
df=df.drop(columns=['Blood Pressure'],axis=1)
df
```

```python
df['Stress Level'].unique()
```

```python
X = df.drop(columns=['Stress Level'],axis=1)
y = df['Stress Level']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.40, random_state=42)
```

```python
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

# **MODELS**

# XGBoost Classifier

```python
from xgboost.sklearn import XGBClassifier

xg_clf = XGBClassifier(n_estimators=500,learning_rate=0.05,max_depth=1)
xg_clf.fit(X_train_scaled, y_train)
print(xg_clf)
```

```python
def print_score_details(clf, X_train_scaled, y_train, X_test_scaled, y_test):
        pred = clf.predict(X_test_scaled)
        clf_report = pd.DataFrame(classification_report(y_test, pred, output_dict=True))
        print("Test Result:\n")        
        print("Accuracy Score:",accuracy_score(y_test, pred),"\n")
        print("Classification Report:\n",clf_report,"\n")
        print("Confusion Matrix:\n",confusion_matrix(y_test, pred),"\n")
```

```python
print_score_details(xg_clf, X_train_scaled, y_train, X_test_scaled, y_test)
```

```python
pred3 = xg_clf.predict(X_train_scaled)
clf_report = pd.DataFrame(classification_report(y_train, pred3, output_dict=True))
print("Train Result:\n")        
print("Accuracy Score:",accuracy_score(y_train, pred3),"\n")
print("Classification Report:\n",clf_report,"\n")
print("Confusion Matrix:\n",confusion_matrix(y_train, pred3),"\n")
```

**Hyperparameter Tuning**

```python
xg_grid = {
    "learning_rate":[0.01, 0.05, 0.1],
    "n_estimators":[100,500,1000],
    "max_depth":[5,10]
}
xg_clf = XGBClassifier()

xg_cv = RandomizedSearchCV(
    xg_clf, 
    xg_grid,
    scoring="f1",
    n_iter=20,
    cv=5
)

xg_cv.fit(X_train_scaled, y_train)
xg_best_params = xg_cv.best_params_
xg_clf = XGBClassifier(**xg_best_params)
xg_clf.fit(X_train_scaled, y_train)
print("Best paramters: ",xg_best_params,"\n")
```

```python
print_score_details(xg_clf, X_train_scaled, y_train, X_test_scaled, y_test)
```

```python
pred3 = xg_clf.predict(X_train_scaled)
clf_report = pd.DataFrame(classification_report(y_train, pred3, output_dict=True))
print("Train Result:\n")        
print("Accuracy Score:",accuracy_score(y_train, pred3),"\n")
print("Classification Report:\n",clf_report,"\n")
print("Confusion Matrix:\n",confusion_matrix(y_train, pred3),"\n")
```

# Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier

tree_clf = DecisionTreeClassifier(random_state=24)
tree_clf.fit(X_train_scaled, y_train)
print_score_details(tree_clf, X_train_scaled, y_train, X_test_scaled, y_test)
```

```python
pred3 = tree_clf.predict(X_train_scaled)
clf_report = pd.DataFrame(classification_report(y_train, pred3, output_dict=True))
print("Train Result:\n")        
print("Accuracy Score:",accuracy_score(y_train, pred3),"\n")
print("Classification Report:\n",clf_report,"\n")
print("Confusion Matrix:\n",confusion_matrix(y_train, pred3),"\n")
```

**The model seems overfit**

```python
from sklearn.model_selection import GridSearchCV

params = {
    "criterion":("gini","entropy"), 
    "splitter":("best", "random"), 
    "min_samples_split":[10,20,30] 
}


tree_cv = GridSearchCV(
    tree_clf, 
    params, 
    scoring="f1", 
    cv=5
) 

tree_cv.fit(X_train_scaled, y_train)
best_params = tree_cv.best_params_
print("Best paramters: ",best_params,"\n")

tree_clf = DecisionTreeClassifier(**best_params)
tree_clf.fit(X_train_scaled, y_train)
print_score_details(tree_clf, X_train_scaled, y_train, X_test_scaled, y_test)
```

```python
pred3 = tree_clf.predict(X_train_scaled)
clf_report = pd.DataFrame(classification_report(y_train, pred3, output_dict=True))
print("Train Result:\n")        
print("Accuracy Score:",accuracy_score(y_train, pred3),"\n")
print("Classification Report:\n",clf_report,"\n")
print("Confusion Matrix:\n",confusion_matrix(y_train, pred3),"\n")
```

**Although the accuracy went down the model is not overfit now.**

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

rf_clf = RandomForestClassifier(n_estimators=300)
rf_clf.fit(X_train_scaled, y_train)
```

```python
print_score_details(rf_clf, X_train_scaled, y_train, X_test_scaled, y_test)
```

```python
predrf = rf_clf.predict(X_train_scaled)
clf_report = pd.DataFrame(classification_report(y_train, predrf, output_dict=True))
print("Train Result:\n")        
print("Accuracy Score:",accuracy_score(y_train, predrf),"\n")
print("Classification Report:\n",clf_report,"\n")
print("Confusion Matrix:\n",confusion_matrix(y_train, predrf),"\n")
```

**The model is overfit**

**Hyperparameter Tuning**

```python
max_features = [5,7]
max_depth = [10,15,20,40]
min_samples_split = [10,30,40]
min_samples_leaf = [2, 10]
bootstrap = [True, False]

random_grid = {
    'max_features': max_features,
    'max_depth': max_depth, 
    'min_samples_split': min_samples_split,
    'min_samples_leaf': min_samples_leaf, 
    'bootstrap': bootstrap
}

rf_cv = RandomizedSearchCV(
    estimator=rf_clf, 
    scoring="f1",
    param_distributions=random_grid, 
    n_iter=20,
    cv=5, 
    
)

rf_cv.fit(X_train_scaled, y_train)
rf_best_params = rf_cv.best_params_
print("Best paramters: ",rf_best_params,"\n")

rf_clf = RandomForestClassifier(**rf_best_params)
rf_clf.fit(X_train_scaled, y_train)

print_score_details(rf_clf, X_train_scaled, y_train, X_test_scaled, y_test)
```

```python
predrf = rf_clf.predict(X_train_scaled)
clf_report = pd.DataFrame(classification_report(y_train, predrf, output_dict=True))
print("Train Result:\n")        
print("Accuracy Score:",accuracy_score(y_train, predrf),"\n")
print("Classification Report:\n",clf_report,"\n")
print("Confusion Matrix:\n",confusion_matrix(y_train, predrf),"\n")
```
